```python
email=input("digite o nome do email que vc gostaria de enviar a mensagem")
imagem=input("qual imagem vc gostaria de pesquisar")
mensagem=input(f"qual mensagem para o email{email} voce gostaria de mandar")

import pyautogui as aut

aut.hotkey("win","r")
aut.sleep(2)
aut.write("chrome") ## abrir chrome
aut.press("ENTER")
aut.sleep(2)
aut.write(imagem)   ## escrever oq gostaria de pesquisar
aut.sleep(2)
aut.press("enter")    #pesquisar
aut.moveTo(372,224)  # mover  em imagens
aut.sleep(2)
aut.click() 
aut.moveTo(130,470) #mover para cima de alguma imagem
aut.sleep(2)
aut.click()  #clicar em alguma imagem 
aut.moveTo(1576,594)
aut.sleep(2)
aut.rightClick()
aut.sleep(2)
aut.moveTo(1639,435) ## mover para "salvar como..."
aut.sleep(5)
aut.click() ## clicar em "salvar como....." 
aut.moveTo(794,498)
aut.sleep(5)
aut.click() # salvar
aut.sleep(5)
aut.moveTo(288,21)## mover para abrir outa paagina
aut.sleep(5)
aut.click()
aut.moveTo(1839,151)## mover para cima do quadrado que contem o email,drive, etc...
aut.sleep(3)
aut.click()
aut.moveTo(1851,257)# mover pra cima do email
aut.sleep(3)
aut.click()#clicar no email
aut.moveTo(127,172)# mover pra cima do composeüìù‚úèÔ∏è
aut.sleep(3)
aut.click()#clicar em compose
aut.moveTo(1429,999)# mover pra cima de anexar arquivosüìéüñáÔ∏è
aut.sleep(3)
aut.click()
aut.sleep(2)

aut.moveTo(349,157)# mover pra cima da pasta que contem a imagem
aut.sleep(3)
aut.click() #clicar na pasta
aut.sleep(2)

aut.moveTo(799,500)# mover mause para cima de "abrir", para abrir a pasta
aut.sleep(3)
aut.click() #clicar em abrir
aut.sleep(2)




aut.moveTo(217,190)# mover para cima de alguma imagem
aut.sleep(3)
aut.click() # clicar na imagem
aut.sleep(2)


aut.moveTo(799,500)# mover mause para cima de "abrir", para anexar a imagem
aut.sleep(3)
aut.click() #clicar em abrir
aut.sleep(2)


aut.moveTo(1295,476)# mover mause para cima de "TO",para escrever pra quem gostaria de enviar o email
aut.sleep(3)
aut.click() #clicar 
aut.sleep(2)
aut.write(email)
aut.press("enter")


aut.moveTo(1329,540)# mover mause para cima de "subject",para escrever uma mensagem para o destinatario
aut.sleep(3)
aut.click() #clicar 
aut.sleep(2)
aut.write(mensagem)

aut.moveTo(1322,1004)# mover mause para enviar mensagem
aut.sleep(3)
aut.click() #clicar 
aut.sleep(2)





```
